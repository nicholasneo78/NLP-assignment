```
import numpy as np
import pandas as pd
import re

COLAB_FILEPATH = './drive/My Drive/4045-assignment-1/'
```

```
data = pd.read_csv(COLAB_FILEPATH+'yelp_Clinton_Street_Baking Company_&_Restaurant_scrapped.csv',encoding='iso-8859-1', header=None)
pd.set_option('display.max_colwidth', -1)
data.head()
```

```
target = [3,4,6,8,11,12,13,14,16,18,20,21,
          25,26,28,33,35,40,41,43,44,49,50,51,52,
          60,61,66,69,72]
for i,j in enumerate(target):
  j -= 2
  target[i] = j

print(target)
```

```
target_data = data.iloc[target]
target_data.head(3)
target_data.iloc[1][1]
```

```
corpus = ""
for i,j in enumerate(target):
  corpus += target_data.iloc[i][1]

corpus[:60]
```

```
# regex the \xa0
txt = str(corpus)
regex = r"\xa0"
txt = re.sub(regex, r" ", txt)
regex = r"\."
txt = re.sub(regex, ". ", txt)
regex = r"([A-Z])"
txt = re.sub(regex, r" \1", txt)
txt.replace("\\", "")
txt = txt.lower()
txt
```

```
# extract the text corpus for future use in 3.3
f = open(COLAB_FILEPATH+"corpus.txt", "w")
f.write(txt)
f.close()
```

```
import nltk 
nltk.download('stopwords')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize, sent_tokenize 
stop_words = set(stopwords.words('english')) 
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```
tokenized = sent_tokenize(txt) 
tagged_list = list()
for i in tokenized: 
      
    # Word tokenizers is used to find the words  
    # and punctuation in a string 
    wordsList = nltk.word_tokenize(i) 
  
    # removing stop words from wordList 
    wordsList = [w for w in wordsList if not w in stop_words]  
  
    #  Using a Tagger. Which is part-of-speech  
    # tagger or POS-tagger.  
    tagged = nltk.pos_tag(wordsList)
    for item in tagged:
      tagged_list.append(item)
    #print(tagged) 

print(tagged_list)
```

```
from operator import itemgetter
```

```
pair_dict = dict()
for i,j in enumerate(tagged_list):
  try:

    if len(tagged_list[i][0]) > 1 and tagged_list[i][1][:2] == "JJ" and tagged_list[i+1][1][:3] == "NNP":
      count = 0
      while tagged_list[i+1+count][1][:3] == "NNP":
        count += 1
      keyname = tagged_list[i][0]
      for i in range(count):
        keyname += "-"+tagged_list[i+1+count][0]
      if keyname[:3] != "css" :
        try:
          pair_dict[keyname] += 1
        except:
          pair_dict[keyname] = 1

    elif len(tagged_list[i][0]) > 1 and tagged_list[i][1][:2] == "JJ" and tagged_list[i+1][1][:2] == "NN":
      keyname = tagged_list[i][0]+"-"+tagged_list[i+1][0]
      if keyname[:3] != "css":
        try:
          pair_dict[keyname] += 1
        except:
          pair_dict[keyname] = 1
  except:
    pass
#pair_dict_sorted = dict(sorted(pair_dict.items(),key=itemgetter(1)))
pair_dict_sorted = sorted(pair_dict.items(),key = lambda x:x[1], reverse=True)
```
