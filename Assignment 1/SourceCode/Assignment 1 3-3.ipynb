```
import numpy as np
import pandas as pd
import re
from textblob import TextBlob

COLAB_FILEPATH = './drive/My Drive/4045-assignment-1/'
```

```
import nltk 
nltk.download('stopwords')
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize, sent_tokenize 
stop_words = set(stopwords.words('english')) 
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
```

```
# read in data
data = pd.read_csv(COLAB_FILEPATH+'corpus.txt',encoding='iso-8859-1', header=None)
pd.set_option('display.max_colwidth', -1)
data = str(data)
data
```

```
regex = r"([!?.])\1{0,}(?=\1)"
data_cleaned = re.sub(regex,'',data)
data_cleaned
```

```
sentences = sent_tokenize(data_cleaned)
len(sentences)
```

```
sentences
```

```
#List for sentences
sentence_library = []
for i in sentences:
  if len(word_tokenize(i)) > 3 and i[:3] != 'css':
    sentence_library.append(i)
    print(word_tokenize(i))
#sentence_library
```

```
len(sentence_library)
```

```
def senti(x):
  if TextBlob(x).sentiment.polarity > 0:
    return "positive", TextBlob(x).sentiment.polarity 
  elif TextBlob(x).sentiment.polarity == 0:
    return "neutral", TextBlob(x).sentiment.polarity 
  else:
    return "negative", TextBlob(x).sentiment.polarity  
```

```
for i,j in enumerate(sentence_library):
  print(j)
  print(senti(sentence_library[i]))
  print()
```
